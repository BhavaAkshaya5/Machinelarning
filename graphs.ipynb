```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.barplot(x="total_bill",y="tip",data=tips)
plt.title("Bar Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("Box Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="tip",data=tips)
plt.title("Violin Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
plt.title("Join Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
iris=sns.load_dataset("iris")
numerical_iris = iris.select_dtypes(include=['number'])
correlation_matrix=numerical_iris.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("correlation Heatmap")
plt.show()
```
